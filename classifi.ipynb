```python
import pandas as pd
df=pd.read_csv('/content/articles.csv')

df
```

```python
df.info()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df["Article_Type"] = label_encoder.fit_transform(df["Article_Type"])

df
```

```python
df.drop(['Id','Article.Banner.Image'],axis=1,inplace=True)

df
```

```python
df['Combined'] = pd.concat([df['Heading'], df['Full_Article']], axis=1).apply(lambda x: ' '.join(x), axis=1)

df
```

```python
x=df['Combined']
y=df['Article_Type']
y
```

```python
from sentence_transformers import SentenceTransformer

model=SentenceTransformer('all-MiniLM-L6-v2')
```

```python
x.to_list()
x
```

```python
embeddings=model.encode(x,batch_size=128,show_progress_bar=True,convert_to_numpy=True)

```

```python
type(embeddings)
```

```python
from sklearn.model_selection import train_test_split
text_train, text_test, sent_train, sent_test = train_test_split(embeddings, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(text_train,sent_train)
```

```python
sent_pred = classifier.predict(text_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

from sklearn import metrics

knn = KNeighborsClassifier(n_neighbors=3)
rfc = RandomForestClassifier(n_estimators = 7,
							criterion = 'entropy',
							random_state =6)
svc = SVC()
lc = LogisticRegression()

for clf in (rfc, knn, svc,lc):
	clf.fit(text_train,sent_train)
	Y_pred = clf.predict(text_test)
	print("Accuracy score of ",
		clf.__class__.__name__,
		"=",100*metrics.accuracy_score(sent_test,
										Y_pred))
```
